```python
def get_icc(results):
    '''get the Intraclass Correlation Coefficient (ICC)'''
    icc = results.cov_re / (results.cov_re + results.scale)
    
    return icc.values[0, 0]
```

```python
model_3 = sm.MixedLM.from_formula("comment_log~like_log+male+female+like_log*male+like_log*female\
                                +model_topic3_ad + model_topic3_ad*male + model_topic3_ad*female", 
                                dat_merge, groups=dat["artist_name_x"])
result_3 = model_3.fit()
print(result_3.summary())
icc_3 = get_icc(result_3)
print("icc: ", icc_3)
```
